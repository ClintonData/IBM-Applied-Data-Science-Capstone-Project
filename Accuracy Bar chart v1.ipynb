```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#1. Define the Accuracy Data
data = {
    'Model': ['Logistic Regression', 'SVM', 'Decision Tree', 'KNN'],
    'Accuracy': [0.8333, 0.8333, 0.7222, 0.8333]
}
```

```python
#2. Create a Pandas DataFrame ---
df = pd.DataFrame(data)
df = df.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)
```

```python
#3. Set Plotting Style
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'sans-serif' # Set a clean font

#4. Create the Bar Chart
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    x='Model', 
    y='Accuracy', 
    data=df, 
    palette='Greens' 
)

#5. Customize the Plot
plt.ylim(0, 1.0) 

# Title and labels
plt.title('Classification Model Accuracy Comparison (Test Set)', fontsize=18, fontweight='bold')
plt.ylabel('Accuracy Score', fontsize=14)
plt.xlabel('Classification Model', fontsize=14)

# Rotate X-axis labels for better readability
plt.xticks(rotation=15, ha='right', fontsize=12)

# Add Accuracy values on top of the bars
for index, row in df.iterrows():
    barplot.text(
        index, 
        row.Accuracy + 0.02, # Position the text slightly above the bar
        f'{row.Accuracy:.4f}', # Format the value to 4 decimal places
        color='black', 
        ha="center",
        fontsize=11,
        fontweight='bold'
    )

# Improve layout
plt.tight_layout()

```

```python
# --- 6. Display the Chart ---
# In a Jupyter Notebook, plt.show() is often redundant, but good practice.
plt.show()

plt.savefig('model_accuracy_comparison.png')
```
